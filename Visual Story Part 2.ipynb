```python
import pandas as pd
from datetime import datetime
df=pd.read_csv('/Users/mandeep/Documents/GitHub/STA380/data/capmetro_UT.csv')
```

```python
import plotly.express as px
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp', inplace=True)
```

```python
df
```

```python
fig = px.box(df, y="boarding", x="hour_of_day",facet_row='weekend')
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
daily_boardings=df[['boarding']].resample('D', kind='period').sum()
daily_temp=df[['temperature']].resample('D', kind='period').mean()
df1=pd.concat([daily_boardings,daily_temp],axis=1)

df1.index = df1.index.strftime('%Y-%m-%d')

# fig = px.line(df1,x=df1.index,y=df1.columns)
# fig.show()

fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(x=df1.index, y=df1['boarding'], name="Boarding"),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=df1.index, y=df1['temperature'], name="Temperature"),
    secondary_y=True,
)
fig.update_layout(
    title_text="Daily boardings and Average Temperatures"
)

# Set x-axis title
fig.update_xaxes(title_text="Dates")

# Set y-axes titles
fig.update_yaxes(title_text="Total Boardings", secondary_y=False)
fig.update_yaxes(title_text="Average Temperature", secondary_y=True)

fig.show()
```

```python
fig = px.box(df, x="boarding", y="day_of_week",category_orders={"day_of_week": ["Mon","Tue","Wed","Thu", "Fri", "Sat", "Sun"]})
fig.show()
```
